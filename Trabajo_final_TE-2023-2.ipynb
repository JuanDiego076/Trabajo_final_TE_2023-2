## Primer punto

#### Ejemplo 4.1 (libro de Sadiku ‚Äì p√°gina: 115): 
Dos cargas puntuales de 1 ùëöùê∂ y ‚àí2 ùëöùê∂ se 
localizan en (3,2, ‚àí 1) y (‚àí1, ‚àí1, 4), respectivamente. Calcule la fuerza el√©ctrica sobre una 
carga de 10 ùëõùê∂ localizada en (0,3, 1) y la intensidad del campo el√©ctrico en ese punto.


Para calcular la fuerza el√©ctrica entre dos cargas puntuales, puedes usar la Ley de Coulomb. La f√≥rmula para la fuerza el√©ctrica ($F$ entre dos cargas $q_1$ y $q_2$) separadas por una distancia ($r$) es:

$ F = k \frac{q_1 q_2}{r^2} $

donde:
- $ F $ es la fuerza el√©ctrica,
- $ k $ es la constante de Coulomb ($8.99 \times 10^9 \ \text{N m}^2/\text{C}^2$)
- $q_1$ y $ q_2 $ son las magnitudes de las cargas,
- $ r $ es la distancia entre las cargas.

Primero, calcularemos la fuerza el√©ctrica entre las dos cargas dadas. Luego, usaremos la f√≥rmula del campo el√©ctrico (\(E\)) para encontrar la intensidad del campo el√©ctrico en un punto debido a una carga:

$$ E = \frac{F}{q} $$

donde:
- $ ( E )$ es la intensidad del campo el√©ctrico,
- $ ( F )$ es la fuerza el√©ctrica,
- $ ( q )$ es la magnitud de la carga en el punto donde se mide el campo el√©ctrico.

Ahora, hagamos los c√°lculos:

1. **Fuerza el√©ctrica entre las cargas:**

$F = k \frac{q_1 q_2}{r^2}$

$ F = (8.99 \times 10^9 \ \text{N m}^2/\text{C}^2) \frac{(1 \times 10^{-3} \ \text{C})(-2 \times 10^{-3} \ \text{C})}{\left(\sqrt{(3-(-1))^2 + (2-(-1))^2 + ((-1)-4)^2}\right)^2} $

2. **Intensidad del campo el√©ctrico en el punto (0, 3, 1):**

$ E = \frac{F}{q} $

$ E = \frac{F}{10^{-9} \ \text{C}} $

Ahora puedes realizar estos c√°lculos para obtener los resultados num√©ricos.

```python
from vpython import *

# Constantes
kel = 8.99e9  # Constante de Coulomb (N m^2 / C^2)

# Cargas
q1 = 1e-3  # 1 mC
q2 = -2e-3  # -2 mC
q_test = 1e-9  # 10 nC

# Posiciones de las cargas
pos_q1 = vector(3, 2, -1)
pos_q2 = vector(-1, -1, 4)
pos_q_test = vector(0, 3, 1)

# Fuerza el√©ctrica entre las cargas
r = pos_q1 - pos_q2
r_magnitude = mag(r)
force_magnitude = kel * q1 * q2 / r_magnitude**2
force_direction = norm(r)
force = force_magnitude * force_direction

# Campo el√©ctrico en el punto debido a la carga de prueba
electric_field_magnitude = force_magnitude / q_test
electric_field_direction = norm(force)
electric_field = electric_field_magnitude * electric_field_direction

# Visualizaci√≥n en VPython
scene = canvas()

# Cargas
charge_q1 = sphere(pos=pos_q1, radius=0.1, color=color.red, charge=q1)
charge_q2 = sphere(pos=pos_q2, radius=0.1, color=color.blue, charge=q2)
charge_test = sphere(pos=pos_q_test, radius=0.05, color=color.green, charge=q_test)

# Flecha que representa la fuerza el√©ctrica
force_arrow = arrow(pos=pos_q_test, axis=force, color=color.yellow, shaftwidth=0.05)

# Flecha que representa el campo el√©ctrico
electric_field_arrow = arrow(pos=pos_q_test, axis=electric_field, color=color.cyan, shaftwidth=0.05)

# Etiquetas
label(pos=pos_q1, text="q1", xoffset=0.5, yoffset=0.5)
label(pos=pos_q2, text="q2", xoffset=0.5, yoffset=0.5)
label(pos=pos_q_test, text="q_test", xoffset=0.5, yoffset=0.5)
```

## Segundo punto

#### Punto 3 (Primer parcial de teor√≠a electromagn√©tica): 
La l√≠nea ùë• = 3, ùëß = ‚àí1, porta una carga 
de 2 ùëõùê∂‚ÅÑùëö, mientras que el plano ùë• = ‚àí2 porta una carga de 4 ùëõùê∂ ùëö2 ‚ÅÑ . Halle la fuerza 
sobre una carga puntual de para ‚àí5 ùëöùê∂ localizada en el origen.

Para calcular la fuerza total sobre la carga puntual en el origen, primero calcularemos las fuerzas debidas a cada distribuci√≥n de carga y luego sumaremos los vectores de fuerza.

1. Fuerza debida a la l√≠nea $x = 3, z = -1$ que porta una carga de $2 \, \text{C/m}$:

$r_1 = \sqrt{(3-0)^2 + (-1-0)^2} = \sqrt{10} \, \text{m} $

$F_1 = \frac{(8.99 \times 10^9 \, \text{N} \cdot \text{m}^2/\text{C}^2) \cdot |(-5 \times 2)|}{(\sqrt{10})^2} $

$F_1 = \frac{8.99 \times 10^9 \, \text{N} \cdot \text{m}^2/\text{C}^2 \cdot 10}{10} = 8.99 \times 10^9 \, \text{N}$

La direcci√≥n de $F_1$ ser√° a lo largo de la l√≠nea $x = 3, z = -1$ y opuesta al vector que conecta la carga puntual en el origen con la l√≠nea.

2. Fuerza debida al plano $x = -2$ que porta una carga de $4 \, \text{C/m}^2$:

$r_2 = |-2 - 0| = 2 \, \text{m} $

$F_2 = \frac{(8.99 \times 10^9 \, \text{N} \cdot \text{m}^2/\text{C}^2) \cdot |(-5 \times 4)|}{2^2} $

$F_2 = \frac{8.99 \times 10^9 \, \text{N} \cdot \text{m}^2/\text{C}^2 \cdot 20}{4} = 4.495 \times 10^{10} \, \text{N}$

La direcci√≥n de $F_2$ ser√° perpendicular al plano $x = -2$ y opuesta al vector que conecta la carga puntual en el origen con el plano.

Ahora, para obtener la fuerza total, sumamos vectorialmente $ F_1 $ y $ F_2 $:

$ \mathbf{F}_{\text{total}} = \mathbf{F}_1 + \mathbf{F}_2 $

La direcci√≥n y magnitud de $ \mathbf{F}_{\text{total}}$ depender√°n de los √°ngulos entre $ \mathbf{F}_1 $ y $ \mathbf{F}_2 $. Puedemos usar la ley de los cosenos para calcular la magnitud y la ley de senos para calcular la direcci√≥n.


```python
from vpython import *

# Constantes
kel = 8.99e9  # Constante de Coulomb (N m^2 / C^2)

# Cargas
q_line = 2e-9  # 2 nC/m
q_plane = 4e-9  # 4 nC/m^2
q_test = -5e-3  # -5 mC

# Distancias
d_line = 1  # Distancia del origen a la l√≠nea
d_plane = 0  # La carga en el plano est√° en x = -2, as√≠ que la distancia es 0

# Funci√≥n para calcular la fuerza
def calculate_force(q1, q2, r1, r2, d):
    return kel * abs(q1 * q2) * log((r2 + sqrt(r2**2 + d**2)) / (r1 + sqrt(r1**2 + d**2)))

# Calcular fuerza desde la carga en la l√≠nea
force_line = calculate_force(q_line, q_test, 3, 3, d_line)

# Calcular fuerza desde la carga en el plano
force_plane = calculate_force(q_plane, q_test, 2, 2, d_plane)

# Posiciones de las cargas
pos_line = vector(3, 0, -1)
pos_plane = vector(-2, 0, 0)
pos_test = vector(0, 0, 0)

# Visualizaci√≥n en VPython
scene = canvas()

# Cargas
charge_line = sphere(pos=pos_line, radius=0.1, color=color.red, charge=q_line)
charge_plane = box(pos=pos_plane, size=vector(0.1, 0.1, 0.1), color=color.blue, charge=q_plane)
charge_test = sphere(pos=pos_test, radius=0.2, color=color.green, charge=q_test)

# Flecha que representa la fuerza total
force_arrow = arrow(pos=pos_test, axis=vector(0, force_line + force_plane, 0), color=color.yellow, shaftwidth=0.05)

# Etiquetas
label(pos=pos_line, text="Carga en l√≠nea", xoffset=0.5, yoffset=0.5)
label(pos=pos_plane, text="Carga en plano", xoffset=0.5, yoffset=0.5)
label(pos=pos_test, text="Carga de prueba", xoffset=0.5, yoffset=0.5)

scene.waitfor('click')  # Esperar a que se haga clic para cerrar la ventana (opcional)
```
